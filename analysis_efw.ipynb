```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data and clean it 
data=pd.read_excel('efw-2019-master-index-data-for-researchers.xlsx',sheet_name='EFW Index 2019 Report',header =4)
data=data.dropna(axis=1,how='all')
pd.set_option('display.max_columns', None)
display(data.head(5))
print("========")
print('data size：{}'.format(data.shape))
print("========")
print('which clo has nan')
print(data.columns[data.isna().any()])
# remove row which has nan value
data=data.dropna(axis=0,how='any')
print("========")
data.describe()
```

```python
#build func
def mov_median(pd_dataframe,country,column,number_of_periods):
    sub_data=pd_dataframe[pd_dataframe.Countries.isin([country])][column]
    if sub_data.shape[0]<number_of_periods:
        return sub_data.median()
    else:
        moving_median=pd.DataFrame(sub_data.iloc[0:number_of_periods,:].median()).T
        for i in range(1,sub_data.shape[0]-number_of_periods):
            moving_median=pd.concat([moving_median, pd.DataFrame(sub_data.iloc[i:i+number_of_periods,:].median()).T])
    return moving_median
```

```python
# test func
value=mov_median(data,'Albania',['Rank','Quartile'],2)
value
```

```python
#preparing data for show first plot
sorted_data=data.sort_values(by='Economic Freedom Summary Index',ascending=False)
display(sorted_data.head(20))
```

```python
sorted_data.head(20).plot.bar(x="Countries",y="Top marginal income and payroll tax rate",title="the TMITR for the top 20 countries that ranked highest in the Economic Freedom Index")
```

```python
panel_data=pd.read_excel('efw-2019-master-index-data-for-researchers.xlsx',sheet_name='EFW Panel Data 2019 Report',header =2)
panel_data=panel_data.dropna(axis=1,how='all')
display(panel_data.head(5))
print("========")
print('data size：{}'.format(panel_data.shape))
print("========")
print('which clo has nan')
print(panel_data.columns[panel_data.isna().any()])
# remove row which has nan value
panel_data=panel_data.dropna(axis=0,how='any')
print("========")
panel_data.describe()
```

```python
moving_median_value=mov_median(panel_data,'United Kingdom',['EFW','Year'],3)
moving_median_value
```

```python
moving_median_value.plot(x='Year',y='EFW',title='the 3-period moving median "EFW" of United Kingdom')
```

## EFW kept rising from 1980 until it peaked in 2000, then began to decline, and remained steady after 2007
